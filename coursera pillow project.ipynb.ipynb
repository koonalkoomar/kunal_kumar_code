```python
import PIL
from PIL import Image
from PIL import ImageEnhance
from PIL import ImageDraw
from PIL import ImageFont
img1= Image.new("RGB", (800, 70), (0,0,0))
image = Image.open('readonly/msi_recruitment.gif').convert('RGB')
imageslist=[]
value=[0.1,0.5,0.9]


# Split into 3 channels

for j in value:
    fnt = ImageFont.truetype("readonly/fanwood-webfont.ttf", 70)
    d = ImageDraw.Draw(img1)
    d.multiline_text((10,10), "channel 0 Intensity {}".format(j), font=fnt, fill=(255, 255, 255))
    #display(img1)
    def get_concat_v(im1, im2):
        dst = Image.new('RGB', (im1.width, im1.height + im2.height))
        dst.paste(im2, (0, 0))
        dst.paste(im1, (0, im2.height))
        return dst
    fimage=get_concat_v(img1, image)
    #display(fimage)
    r, g, b = fimage.split()
    r = r.point(lambda i: i * j )
    g = g.point(lambda i: i * 1)
    b = b.point(lambda i: i * 1)

    result1 = Image.merge('RGB', (r, g, b))
    result1.save('result1.png')
    #display(result1)
    imageslist.append(result1)
    

    
    
    
    
for j in value:
    img1 = Image.new("RGB", (800, 70), (0,0,0))
    image = Image.open('readonly/msi_recruitment.gif').convert('RGB')
    
    fnt = ImageFont.truetype("readonly/fanwood-webfont.ttf", 40)
    d = ImageDraw.Draw(img1)
    d.multiline_text((10,10), "channel 1 Intensity {}".format(j), font=fnt, fill=(255, 255, 255))
    #display(img1)
    def get_concat_v(im1, im2):
        dst = Image.new('RGB', (im1.width, im1.height + im2.height))
        dst.paste(im2, (0, 0))
        dst.paste(im1, (0, im2.height))
        return dst
    fimage=get_concat_v(img1, image)
    #display(fimage)
    r, g, b =fimage.split()
    r = r.point(lambda i: i * 1)
    g = g.point(lambda i: i * j)
    b = b.point(lambda i: i * 1)

    result2 = Image.merge('RGB', (r, g, b))
    result2.save('result2.png')
    #display(result2)
    imageslist.append(result2)
    

for j in value:
    img1 = Image.new("RGB", (800, 70), (0,0,0))
    image = Image.open('readonly/msi_recruitment.gif').convert('RGB')
    
    fnt = ImageFont.truetype("readonly/fanwood-webfont.ttf", 40)
    d = ImageDraw.Draw(img1)
    d.multiline_text((10,10), "channel 2 Intensity {}".format(j), font=fnt, fill=(255, 255, 255))
    #display(img1)
    def get_concat_v(im1, im2):
        dst = Image.new('RGB', (im1.width, im1.height + im2.height))
        dst.paste(im2, (0, 0))
        dst.paste(im1, (0, im2.height))
        return dst
    fimage=get_concat_v(img1, image)
    #display(fimage)
    r, g, b = fimage.split()
    r = r.point(lambda i: i * 1 )
    g = g.point(lambda i: i * 1)
    b = b.point(lambda i: i * j)

    result3 = Image.merge('RGB', (r, g, b))
    result3.save('result3.png')
    #display(result3)
    imageslist.append(result3)
    
print(imageslist)
first_image=imageslist[0]
contact_sheet=PIL.Image.new(first_image.mode, (first_image.width*3,first_image.height*3))
x=0
y=0

# build a list of 9 images which have different brightnesses


for img in imageslist:
    # Lets paste the current image into the contact sheet
    contact_sheet.paste(img, (x, y) )
    # Now we update our X position. If it is going to be the width of the image, then we set it to 0
    # and update Y as well to point to the next "line" of the contact sheet.
    if x+first_image.width == contact_sheet.width:
        x=0
        y=y+first_image.height
    else:
        x=x+first_image.width

# resize and display the contact sheet
contact_sheet = contact_sheet.resize((int(contact_sheet.width/2),int(contact_sheet.height/2) ))
display(contact_sheet)
```
